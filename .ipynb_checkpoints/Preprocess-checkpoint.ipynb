```python
import scanpy as sc
import pandas as pd
import scvi
import numpy as np
```

Load Files for Plate 7,8,9

```python
plate7 = sc.read_h5ad('tahoe_data/plate7_DMSO_Dinaciclib_A549_HT-29.h5ad')
plate8 = sc.read_h5ad('tahoe_data/plate8_DMSO_Dinaciclib_A549_HT-29.h5ad')
plate9 = sc.read_h5ad('tahoe_data/plate9_DMSO_Dinaciclib_A549_HT-29.h5ad')
```

```python
adata_78 = plate7.concatenate(plate8)
adata = adata_78.concatenate(plate9)

sc.pp.sample(adata,n = 1000)
```

```python
print(adata.X.shape)               # cells × genes
print(adata.obs.columns.tolist()) # metadata columns for cells
print(adata.var.columns.tolist()) # metadata columns for genes
print(adata.layers.keys())        # available layers
print(adata.uns.keys())           # unstructured annotations
```

```python
adata

adata.layers["counts"] = adata.X.copy()  # preserve counts

```

```python

cell_sums = adata.X.sum(axis=1)
print(f"Per-cell total (first 10): {cell_sums[:10]}")

# Heuristic: if median ≈ mean ≈ constant, it's likely normalized
print(f"Min: {np.min(cell_sums)}, Max: {np.max(cell_sums)}, Mean: {np.mean(cell_sums):.2f}, Std: {np.std(cell_sums):.2f}")
```

```python
scvi.model.SCVI.setup_anndata(
    adata,
    layer="counts",
    continuous_covariate_keys=["pcnt_mito"],
    batch_key="batch"
)
```

```python
model = scvi.model.SCVI(adata)
```

```python
model
```

```python
model.train(max_epochs = 50)
```

```python
SCVI_LATENT_KEY = "X_scVI"

latent = model.get_latent_representation()
adata.obsm[SCVI_LATENT_KEY] = latent
latent.shape
```

```python
# use scVI latent space for UMAP generation
sc.pp.neighbors(adata, use_rep=SCVI_LATENT_KEY)
sc.tl.umap(adata, min_dist=0.1)

sc.pl.umap(
    adata,
    color=["cell_name"],
    frameon=False,
)
sc.pl.umap(
    adata,
    color=["phase", "drugconc"],
    ncols=2,
    frameon=False,
)
```

```python
# run PCA then generate UMAP plots
sc.tl.pca(adata)
sc.pp.neighbors(adata, n_pcs=30, n_neighbors=20)
sc.tl.umap(adata, min_dist=0.1)
```

```python
sc.pl.umap(
    adata,
    color=["cell_name"],
    frameon=False,
)
sc.pl.umap(
    adata,
    color=["phase", "drugconc"],
    ncols=2,
    frameon=False,
)
```
