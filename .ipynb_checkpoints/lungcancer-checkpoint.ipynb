```python
import scanpy as sc
import pandas as pd

# Ensure "cell_name" contains only SHP-77 and SW 1271
adata_subset = adata[adata.obs["cell_name"].isin(["SHP-77", "A549"])].copy()

# Run DE: SHP-77 vs SW 1271 (SHP-77 as group, SW 1271 as reference)
sc.tl.rank_genes_groups(
    adata_subset,
    groupby="cell_name",
    groups=["SHP-77"],
    reference="A549",
    method="wilcoxon",
    use_raw=False  # Use normalized expression in adata.X
)

# Extract DE results
de_result = pd.DataFrame({
    "gene": adata_subset.uns["rank_genes_groups"]["names"]["SHP-77"],
    "logfoldchanges": adata_subset.uns["rank_genes_groups"]["logfoldchanges"]["SHP-77"],
    "pvals": adata_subset.uns["rank_genes_groups"]["pvals"]["SHP-77"],
    "pvals_adj": adata_subset.uns["rank_genes_groups"]["pvals_adj"]["SHP-77"],
    "scores": adata_subset.uns["rank_genes_groups"]["scores"]["SHP-77"]
})

# View top results
print(de_result.head(200))
```

```python
sig_genes = de_result["pvals_adj"] < 0.05
percentage = sig_genes.mean() * 100
print(f"{percentage:.2f}% of genes are significant at adjusted p < 0.05")
```

```python
import scanpy as sc
import pandas as pd

phases = ["G1", "S", "G2M"]
results = []

for phase in phases:
    # Subset to cells in the current phase and in SHP-77 or SW 1271
    adata_phase = adata[
        (adata.obs["phase"] == phase) &
        (adata.obs["cell_name"].isin(["SHP-77", "A549"]))
    ].copy()

    # Run DE between SHP-77 and SW 1271 in this phase
    sc.tl.rank_genes_groups(
        adata_phase,
        groupby="cell_name",
        groups=["SHP-77"],
        reference="A549",
        method="wilcoxon",
        use_raw=False
    )

    # Extract stats for TOP2A
    names = adata_phase.uns["rank_genes_groups"]["names"]["SHP-77"]
    pvals_adj = adata_phase.uns["rank_genes_groups"]["pvals_adj"]["SHP-77"]
    lfc = adata_phase.uns["rank_genes_groups"]["logfoldchanges"]["SHP-77"]
    scores = adata_phase.uns["rank_genes_groups"]["scores"]["SHP-77"]

    if "TOP2A" in names:
        idx = list(names).index("TOP2A")
        results.append({
            "phase": phase,
            "logfoldchange": lfc[idx],
            "pval_adj": pvals_adj[idx],
            "score": scores[idx]
        })
    else:
        results.append({
            "phase": phase,
            "logfoldchange": None,
            "pval_adj": None,
            "score": None
        })

# Display results
df_results = pd.DataFrame(results)
print(df_results)
```

```python
# List of cell cycle genes of interest
genes = ["TOP2A", "MKI67", "CDK1", "CCNB1", "PCNA"]

# Initialize results
summary = []

for gene in genes:
    row = {"gene": gene}

    # --- Combined DE ---
    if gene in de_result["gene"].values:
        idx = de_result["gene"] == gene
        row["combined_padj"] = de_result.loc[idx, "pvals_adj"].values[0]
    else:
        row["combined_padj"] = None

    # --- Per-phase DE ---
    for _, r in df_results.iterrows():
        phase = r["phase"]
        if r["phase"] == phase and r["gene"] == gene if "gene" in r else False:
            row[f"{phase}_padj"] = r["pval_adj"]
        else:
            # fallback: if df_results has only one gene (e.g. TOP2A)
            if gene == "TOP2A" and "phase" in r:
                row[f"{r['phase']}_padj"] = r["pval_adj"]

    summary.append(row)

# Create and display DataFrame
summary_df = pd.DataFrame(summary)
print(summary_df)
```

```python

# --- DE: combined across all phases ---
sc.tl.rank_genes_groups(
    adata_subset,
    groupby="cell_name",
    groups=["SHP-77"],
    reference="A549",
    method="wilcoxon",
    use_raw=False
)

# Extract full DE result
de_combined = pd.DataFrame({
    "gene": adata_subset.uns["rank_genes_groups"]["names"]["SHP-77"],
    "logfc_combined": adata_subset.uns["rank_genes_groups"]["logfoldchanges"]["SHP-77"],
    "padj_combined": adata_subset.uns["rank_genes_groups"]["pvals_adj"]["SHP-77"],
})

# --- DE per phase ---
phases = ["G1", "S", "G2M"]
phase_dfs = []

for phase in phases:
    adata_phase = adata_subset[adata_subset.obs["phase"] == phase].copy()

    sc.tl.rank_genes_groups(
        adata_phase,
        groupby="cell_name",
        groups=["SHP-77"],
        reference="A549",
        method="wilcoxon",
        use_raw=False
    )

    df = pd.DataFrame({
        "gene": adata_phase.uns["rank_genes_groups"]["names"]["SHP-77"],
        f"logfc_{phase}": adata_phase.uns["rank_genes_groups"]["logfoldchanges"]["SHP-77"],
        f"padj_{phase}": adata_phase.uns["rank_genes_groups"]["pvals_adj"]["SHP-77"]
    })

    phase_dfs.append(df)

# --- Merge all results on gene ---
from functools import reduce

all_results = reduce(lambda left, right: pd.merge(left, right, on="gene", how="outer"), [de_combined] + phase_dfs)

# --- Filter to 20 canonical cell cycle genes ---
cell_cycle_genes = [
    "TOP2A", "MKI67", "CDK1", "CCNB1", "PCNA", "BUB1", "CDC20", "AURKB", "CENPF", "UBE2C",
    "NUSAP1", "TYMS", "RRM2", "CCNA2", "CDT1", "GMNN", "E2F1", "EXO1", "KIF11", "PLK1"
]

cc_results = all_results[all_results["gene"].isin(cell_cycle_genes)].reset_index(drop=True)

# --- Display or export ---
print(cc_results)
```
