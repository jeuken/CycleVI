```python
import pandas as pd
import anndata as ad
import scanpy as sc
import matplotlib.pyplot as plt
from sklearn.preprocessing import QuantileTransformer
import numpy as np

# Load files
expr = pd.read_csv("/Users/piamozdzanowski/Documents/GitHub/PerturbCycleVI/fucci_expression.csv", index_col=0)
meta = pd.read_csv("/Users/piamozdzanowski/Documents/GitHub/PerturbCycleVI/fucci_metadata.csv")

# Filter gene columns to those starting with "ENSG"
ensg_genes = [gene for gene in expr.columns if gene.startswith("ENSG")]
expr = expr[ensg_genes]

# Align metadata
meta = meta.set_index("cell_id")
meta = meta.loc[expr.index]  # ensure cell IDs match

# Create AnnData object
adata = ad.AnnData(X=expr.values, obs=meta, var=pd.DataFrame(index=expr.columns))

# Add raw counts as layer
adata.layers["counts"] = adata.X.copy()
```

```python
import numpy as np

# Check if values are integers (typical for raw counts)
are_integers = np.all(np.equal(np.mod(expr.values, 1), 0))

# Check if any values are negative (they shouldn't be in count data)
has_negatives = np.any(expr.values < 0)

# Quick summary
print(f"All integers: {are_integers}")
print(f"Any negatives: {has_negatives}")
print(f"Value range: min={expr.values.min()}, max={expr.values.max()}")
```

```python
# Load gene-phase mapping (with Ensembl IDs and 'phase' column)
cell_cycle_genes = pd.read_csv("/Users/piamozdzanowski/Documents/GitHub/PerturbCycleVI/benchmark_data/Homo_sapiens.csv")

# Extract gene lists by phase
g2m_genes = cell_cycle_genes.query("phase == 'G2/M'")["geneID"].unique().tolist()
s_genes = cell_cycle_genes.query("phase == 'S'")["geneID"].unique().tolist()
```

```python
sc.pp.normalize_total(adata)
sc.pp.log1p(adata)
sc.pp.scale(adata)

# Compute cell cycle scores (if not already done)
sc.tl.score_genes_cell_cycle(adata, s_genes=s_genes, g2m_genes=g2m_genes)
```

```python
score_angle = np.arctan2(adata.obs["G2M_score"], adata.obs["S_score"])
score_angle = np.mod(score_angle, 2 * np.pi)  # wrap to [0, 2π]

# Store in obs
adata.obs["cycle_angle"] = score_angle
```

```python
def quantile_to_uniform_2pi(adata, angle_key, output_key):
    n_cells = adata.n_obs
    qt = QuantileTransformer(
        output_distribution='uniform',
        random_state=0,
        n_quantiles=n_cells,
        subsample=n_cells
    )
    angle_array = adata.obs[angle_key].values.reshape(-1, 1)
    
    # Handle NaNs gracefully
    if np.isnan(angle_array).any():
        raise ValueError(f"{angle_key} contains NaNs. Please clean or impute missing values.")

    angle_uniform = qt.fit_transform(angle_array).flatten()
    adata.obs[output_key] = angle_uniform * 2 * np.pi

quantile_to_uniform_2pi(adata, angle_key="cycle_angle", output_key="cycle_angle_uniform")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract angles
angle_raw = adata.obs["cycle_angle"]
angle_uniform = adata.obs["cycle_angle_uniform"]

# Define bins over [0, 2π]
bins = np.linspace(0, 2 * np.pi, 50)

# Create histograms
fig, axs = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

# Raw angle histogram
axs[0].hist(angle_raw, bins=bins, color='steelblue', alpha=0.7)
axs[0].set_title("Original Cycle Angle Distribution")
axs[0].set_xlabel("Angle (radians)")
axs[0].set_ylabel("Cell Count")
axs[0].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[0].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])
axs[0].grid(True)

# Uniform-transformed angle histogram
axs[1].hist(angle_uniform, bins=bins, color='darkorange', alpha=0.7)
axs[1].set_title("Uniform-Transformed Angle Distribution")
axs[1].set_xlabel("Angle (radians)")
axs[1].set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
axs[1].set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

# Extract values
theta = adata.obs["theta"]
gfp = adata.obs["gfp.median.log10sum.adjust"]
rfp = adata.obs["rfp.median.log10sum.adjust"]
dapi = adata.obs["dapi.median.log10sum.adjust"]
size = adata.obs["size"]
# Sort theta for smooth plotting
sort_idx = np.argsort(theta)
theta_sorted = theta[sort_idx]

# Apply LOWESS smoothing
gfp_smooth = sm.nonparametric.lowess(gfp, theta, frac=0.3, return_sorted=False)
rfp_smooth = sm.nonparametric.lowess(rfp, theta, frac=0.3, return_sorted=False)
size_smooth = sm.nonparametric.lowess(size, theta, frac=0.3, return_sorted=False)

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 8), sharex=True)

# Plot fluorescence with smooth curves
axs[0].scatter(theta, gfp, color='green', alpha=0.4, label='GFP', s=10)
axs[0].scatter(theta, rfp, color='red', alpha=0.4, label='RFP', s=10)
axs[0].plot(theta_sorted, gfp_smooth[sort_idx], color='green', linewidth=2, label='GFP (LOWESS)')
axs[0].plot(theta_sorted, rfp_smooth[sort_idx], color='red', linewidth=2, label='RFP (LOWESS)')
axs[0].set_ylabel("Fluorescence (log10 adjusted)")
axs[0].set_title("FUCCI Fluorescence vs. Theta")
axs[0].legend()
axs[0].grid(True)

# Plot cell size with smooth curve
axs[1].scatter(theta, size, color='gray', alpha=0.4, label='Cell size', s=10)
axs[1].plot(theta_sorted, size_smooth[sort_idx], color='black', linewidth=2, label='Size (LOWESS)')
axs[1].set_xlabel("Theta (cell cycle phase)")
axs[1].set_ylabel("Cell size")
axs[1].set_title("Cell Size vs. Theta")
axs[1].grid(True)

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
from mygene import MyGeneInfo

# Load gene list
gene_symbols = pd.read_csv("cluster_files/cell_cycle_genes.txt", header=None, names=["symbol"])["symbol"].tolist()

# Initialize mygene
mg = MyGeneInfo()

# Query Ensembl Gene IDs for human gene symbols
result = mg.querymany(gene_symbols, scopes='symbol', fields='ensembl.gene', species='human')

# Extract only valid Ensembl gene IDs
ensembl_ids = []
for r in result:
    if 'ensembl' in r:
        ensg = r['ensembl']
        if isinstance(ensg, list):
            ensembl_ids.append(ensg[0]['gene'])  # take the first if multiple
        else:
            ensembl_ids.append(ensg['gene'])

# Remove duplicates and save as .txt
with open("cluster_files/cell_cycle_genes_ensembl.txt", "w") as f:
    for eid in sorted(set(ensembl_ids)):
        f.write(f"{eid}\n")
```
